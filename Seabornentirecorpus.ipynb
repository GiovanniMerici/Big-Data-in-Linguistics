```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
#ELEMENT PREPOSITIONNEL DONE
dfFRpo = pd.read_excel ('FRANTEXTpourque.xlsx') 
dfBFMpo = pd.read_excel ('BFMpourque.xlsx')
dfFRpa = pd.read_excel ('FRANTEXTparceque.xlsx')
dfBFMpa = pd.read_excel ('BFMparceque.xlsx')

#ELEMENT ADVERBIAL DONE
dfBFMta = pd.read_excel('BFMTandisQue.xlsx')
dfFRta = pd.read_excel ('FRANTEXTTandisQue.xlsx')
dfBFMau = pd.read_excel ('BFMAussitotQue.xlsx')
dfFRau = pd.read_excel ('FRANTEXTAussitotQue.xlsx')

#ELEMENT PARTICIPIAL DONE
dfBFMmo = pd.read_excel('BFMMoyennantQue.xlsx')
dfFRmo = pd.read_excel ('FRANTEXTMoyennantQue.xlsx')
dfFRat = pd.read_excel ('FRANTEXTAttenduQue.xlsx')

#ELEMENT NOMINAL
dfBFMco = pd.read_excel('BFMAConditionQue.xlsx')
dfFRco = pd.read_excel ('FRANTEXTAConditionQue.xlsx')
dfBFMme = pd.read_excel ('BFMAMesureQue.xlsx')
dfFRme = pd.read_excel ('FRANTEXTAMesureQue.xlsx') 
```

```python
dfBFMpo1 = dfBFMpo.drop(columns=['LC', 'RC'])
dfFRpo1 = dfFRpo.drop(columns=['ID', 'LC', 'RC'])
dfBFMpa1 = dfBFMpa.drop(columns=['LC', 'RC'])
dfFRpa1 = dfFRpa.drop(columns=['ID', 'LC', 'RC'])

dfBFMta1 = dfBFMta.drop(columns=['LC', 'RC'])
dfFRta1 = dfFRta.drop(columns=['ID', 'LC', 'RC'])
dfBFMau1 = dfBFMau.drop(columns=['LC', 'RC'])
dfFRau1 = dfFRau.drop(columns=['ID', 'LC', 'RC'])

dfBFMmo1 = dfBFMmo.drop(columns=['LC', 'RC'])
dfFRmo1 = dfFRmo.drop(columns=['ID', 'LC', 'RC'])
dfFRat1 = dfFRat.drop(columns=['ID', 'LC', 'RC'])

dfBFMco1 = dfBFMco.drop(columns=['LC', 'RC'])
dfFRco1 = dfFRco.drop(columns=['ID', 'LC', 'RC'])
dfBFMme1 = dfBFMme.drop(columns=['LC', 'RC'])
dfFRme1 = dfFRme.drop(columns=['ID', 'LC', 'RC'])
```

ELEMENTS PREPOSITIONNELS

```python
sns.set(rc={'figure.figsize':(20,10)})
#'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen'

sns.set_color_codes()
sns.distplot(dfBFMpo1['Date'], color='darkblue').set(title="Evolution de 'pour que' d'après les données de la BFM et de Frantext")
sns.distplot(dfFRpo1['Date'], color='blue')
sns.rugplot(dfBFMpo1['Date'], color='darkblue')
sns.rugplot(dfFRpo1['Date'], color='blue')


plt.legend(labels=["Pour que BFM", "Pour que Frantext"], ncol=2, title ="Légende", fontsize = 'large', title_fontsize = "15", loc = 2, bbox_to_anchor = (0,1))
plt.xlim(1100,1799)
plt.xticks(np.arange(1100,1799,50))
#plt.savefig('seabornPourQue.png', dpi=1080)
```
